```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(float(1000))
y = x

plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(float(1000))
y = x

fig, ax = plt.subplots()
ax.plot(x, y)
ax.plot(x, x**2,color='r')

plt.show()
```
